```python
welcome_message = f"""Welcome to TwerkQueen Company! We are excited to have you on board. 
Your journey is important to us, and we are committed to ensuring you have a safe and pleasant flight. 
Thank you for choosing TwerkQueen Company, and we look forward to serving you!
"""

print(welcome_message) 

from passenger_personal_information import get_input

import pandas as pd

df = pd.read_csv("thailand_flights_with_updated_airlines.csv")

available_from_cities = df["From City"].unique()
available_to_cities = df["To City"].unique()

print("Available provinces for origin:")
print(available_from_cities)

print("\nAvailable provinces for destination:")
print(available_to_cities)

df.head()

def find_flights(from_city, to_city):
    filtered_flights = df[(df["From City"] == from_city) & (df["To City"] == to_city)]
    sorted_flights = filtered_flights.sort_values(by="Price (Baht)")
    return sorted_flights

from_city = str(input("Please enter your province of origin : "))
to_city = str(input("Please enter your destination province : "))
result = find_flights(from_city, to_city)

result



```
